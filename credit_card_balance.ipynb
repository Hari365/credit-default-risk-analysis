```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit = pd.read_csv('credit_card_balance.csv')
```

```python
credit.shape
```

```python
credit.head(10).T
```

```python
credit = pd.concat([credit, pd.get_dummies(credit['NAME_CONTRACT_STATUS'])], axis=1).drop('NAME_CONTRACT_STATUS', axis=1)
```

```python
credit.isnull().sum()
```

```python
credit['AMT_DRAWINGS_ATM_CURRENT'].fillna(credit['AMT_DRAWINGS_ATM_CURRENT'].mean(), inplace=True)
```

```python
credit['AMT_DRAWINGS_OTHER_CURRENT'].fillna(credit['AMT_DRAWINGS_OTHER_CURRENT'].mean(), inplace=True)
```

```python
credit['AMT_DRAWINGS_POS_CURRENT'].fillna(credit['AMT_DRAWINGS_POS_CURRENT'].mean(), inplace=True)
```

```python
credit['AMT_INST_MIN_REGULARITY'].fillna(credit['AMT_INST_MIN_REGULARITY'].mean(), inplace=True)
```

```python
credit['AMT_PAYMENT_CURRENT'].fillna(credit['AMT_PAYMENT_CURRENT'].mean(), inplace=True)
```

```python
credit['CNT_DRAWINGS_ATM_CURRENT'].fillna(credit['CNT_DRAWINGS_ATM_CURRENT'].mean(), inplace=True)
```

```python
credit['CNT_DRAWINGS_OTHER_CURRENT'].fillna(credit['CNT_DRAWINGS_OTHER_CURRENT'].mean(), inplace=True)
```

```python
credit['CNT_DRAWINGS_POS_CURRENT'].fillna(credit['CNT_DRAWINGS_POS_CURRENT'].mean(), inplace=True)
```

```python
credit['CNT_INSTALMENT_MATURE_CUM'].fillna(credit['CNT_INSTALMENT_MATURE_CUM'].mean(), inplace=True)
```

```python
credit_app_curr = credit.groupby('SK_ID_CURR').mean()
```
