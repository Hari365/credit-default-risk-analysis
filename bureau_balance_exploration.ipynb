```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
```

```python
bureau_balance = pd.read_csv('bureau_balance.csv')
```

```python
bureau_balance.head()
```

```python
fig = plt.figure(figsize=(24,10))
miss = pd.DataFrame((bureau_balance.isnull().sum()*100)/bureau_balance.shape[0]).reset_index()
ax = sns.pointplot(x='index', y=0, data=miss)
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
bureau_balance['STATUS'].value_counts()
```

```python
plt.figure(figsize=(10,10))
bureau_balance['STATUS'].value_counts().plot.pie(autopct='%1.00f%%')
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.distplot(bureau_balance['MONTHS_BALANCE'])
plt.show()
```
